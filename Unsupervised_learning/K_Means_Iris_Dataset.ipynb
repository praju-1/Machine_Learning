```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], df['petal length (cm)'], df['petal width (cm)'])
plt.xlabel('Sepal Data')
plt.ylabel('Petal Data')
plt.title('Scatter plot')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit_predict(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])
```

```python
df['clusters'] = model.fit_predict(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])
```

```python
df.head()
```

```python
df0 = df[df['clusters'] == 0]
df1 = df[df['clusters'] == 1]
df2 = df[df['clusters'] == 2]
```
