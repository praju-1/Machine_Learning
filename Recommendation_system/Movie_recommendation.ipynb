```python
combined_feature = df1['genres'] + " " + df1['keywords'] + " " + df1['tagline'] + " " +  df1['cast'] + " " + df1['director']
```

```python
combined_feature[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectors = CountVectorizer()
```

```python
feature_vectors = vectors.fit_transform(combined_feature)
```

```python
import difflib
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
final_match = list(enumerate(similarity[index_of_the_movie]))
```

```python
final_match 
```

```python
from operator import itemgetter
```

```python
sorted_match = sorted(final_match, key = itemgetter(1,0), reverse=True)
```

```python
sorted_match
```

- some for loop techniques by which we can access index and its score of similarity

```python
n = 0
for i in sorted_match:
    n += 1
    if n<6:
        print(sorted_match[n])
```

```python
for i in range(1,6):
    print(sorted_match[i])
```

- Below we get only index value for the similarity recommendation of the movies

```python
print("Movie suggested for you : ")

for i in range(1,6):
    index = sorted_match[i][0]
    print(index)
```
