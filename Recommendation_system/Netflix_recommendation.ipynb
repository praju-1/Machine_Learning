```python
def show_category(movie):
    index_of_the_show = df[df["title"] == movie].index[0]
    show_type = df.loc[index_of_the_show, "type"]
    return show_type

show_type = show_category('Vendetta: Truth, Lies and The Mafia')
if show_type:
    print(f"The show type is: {show_type}")
```

```python
from operator import itemgetter

def recommend_and_show_category(movie):
    # Find the index of the given movie title in the movie DataFrame
    try:
        movie_index = df1[df1["title"] == movie].index[0]
        
        # Get the recommended movies based on similarity
        dist = similarity[movie_index]
        movie_list = sorted(list(enumerate(dist)), key=itemgetter(1,0), reverse=True)[:6]

        print("Recommended movies:")
        for i in movie_list:
            recommended_movie_title = df1.iloc[i[0]].title
            print(recommended_movie_title)

        # Get the category (type) of the given movie
        index_of_the_show = df[df["title"] == movie].index[0]
        show_type = df.loc[index_of_the_show, "type"]
        print(f"\nThe show type of '{movie}' is: {show_type}")

    except IndexError:
        print("Movie not found or insufficient data.")
```

```python
# Example usage
recommend_and_show_category('Vendetta: Truth, Lies and The Mafia')
```

```python
recommend_and_show_category('Dick Johnson Is Dead')
```
